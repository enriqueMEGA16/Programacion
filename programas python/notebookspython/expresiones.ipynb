# Operaciones

En esta area se realizan las operaciones de Aritmetica `basicas`


````python
suma = 4 + 2
resta = 4-2
multi = 4*2
div = 10 / 3
divEntera = 10 // 3
mod = 10 % 2
potencia = 10 ** 2
salida = f"""==== Resultados de Operadores Aritmeticos
\nLa suma es: {suma}
\n La resta es: {resta}
\n La multiplicacion es: {multi}
\n La division es: {div}
\n La division entera es: {divEntera}
\n El modulo es: {mod}
\n La potencia es: {potencia}
"""
print(salida)
````
